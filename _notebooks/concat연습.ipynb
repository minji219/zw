```python
import pandas as pd
```

```python
df2016=pd.read_csv("https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/stocks_2016.csv")
df2017=pd.read_csv("https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/stocks_2017.csv")
df2018=pd.read_csv("https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/stocks_2018.csv")
```

concat은 음 약간 R에서 rbind cbind같은 느낌..

```python
df2016
```

```python
df2017
```

```python
df2018
```

```python
pd.concat([df2016,df2017,df2018])  #concat으로 합치기  인덱스가 좀 이상하다. 
```

```python
pd.concat([df2016,df2017,df2018]).reset_index()  #인덱스ㅜ 다시 설정
```

```python
pd.concat([df2016,df2017,df2018]).reset_index(drop=True)  #reset_index안에 drop=True설정하여 원래 인덱스 없앰
```

R에서 rbind와 유사 

`-`rbind와의 차이점 : 합치려는 데이터 프레임의 columns가 꼭 동일할필요는 없다. 

```python
df2017
```

```python
df2017.iloc[:,1:]  #세로
```

```python
df2017.iloc[1:,:] #가로
```

```python
pd.concat([df2016,df2017.iloc[:,1:]])  #columns가 달라도 상관 없음 걍 nan으로 들어감
```

#### pd.concat([...],axis='columns')

R에서 cbind랑 비슷한 느낌 근데 row의 숫자가 서로 달라도 괜찮은...

```python
df2016
```

```python
df2017
```

```python
df2018
```

```python
pd.concat([df2016,df2017,df2018],axis='columns')
```

```python
pd.concat([df2016,df2017,df2018],axis=1)  #'columns'를 1로 해도 괜찮
```

#### pd.concat([...],keys=[...])

```python
pd.concat([df2016,df2017,df2018],keys=[2016,2017,2018])  # 인덱스를 나눠줌..? 제목을 붙혀줌...이런맥락
```

`-`위의 코드는 아래와 같다. 

```python
pd.concat({2016:df2016,2017:df2017,2018:df2018})
```

```python
pd.concat({df2016:2016,df2017:2017,df2018:2018})  #이건안됨
```

```python
pd.concat(dict(zip([2016,2017,2018],[df2016,df2017,df2018]))) #zip으로 묶기
```

`-`의도하지않았는데 그냥 되는것들..

```python
pd.concat([df2016,df2017,df2018],keys=[2016,2017])
```

그럼걍 2016,2017 꺼만 나옴

```python
pd.concat([df2016,df2017,df2018],keys=[2017,2018]) #2016꺼가 빠짐
```

#### pd.concat([...],keys=[...],axis='columns')

```python
pd.concat([df2016,df2017,df2018],keys=[2016,2017,2018],axis=1)
```

```python
pd.concat({2016:df2016,2017:df2017,2018:df2018},axis=1)
```

```python
pd.concat(dict(zip([2016,2017,2018],[df2016,df2017,df2018])),axis=1)
```
